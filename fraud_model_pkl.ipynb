importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# first five row of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset information
credit_card_data.info()
```

```python
#checking the number missing value in each colum
credit_card_data.isnull().sum()
```

```python
#distribution of legit transaction and fradulent transaction
credit_card_data['Class'].value_counts()
```

this data is highly unbalanced

0->normal transaction
1-> fraudulent transaction

```python
#seperating the data for analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measure of the data
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the value for both transaction
credit_card_data.groupby('Class').mean()
```

under-sampling
build a sample data containing similar distribution of normal transaction and fraudulent transaction


```python
legit_sample=legit.sample(n=492)
```

contacting two data frame


```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class',axis=1)
Y= new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

split the data into training data and testing data  

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression()
```

```python
# training the logistics regression model with training data
model.fit(X_train,Y_train)
```

```python
#model evaluation
```

```python
# accuracy on traing data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('accurancy on traing data:',training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy scrore on test data : ', test_data_accuracy)
```
